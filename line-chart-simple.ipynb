# Simple line chart

## Import Modules

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-v0_8')
```

## Testdata

```python
# x-data with 200 points
x = np.linspace(0, 5, 200)  

y1 = 10 * x     # curve 1
y2 = 3*x**2     # curve 2
y3 = x**3       # curve 3
```

## Simple chart

```python
# create objects figure und axes, size: 9x4 inch
fig, ax = plt.subplots(figsize = (9,4))

# plot all curves with labels for legend
ax.plot(x, y1, label = 'linear')
ax.plot(x, y2, label = 'square')
ax.plot(x, y3, label = 'cubic')

ax.set_xlabel('x-value')        # Add x-label to the axes
ax.set_ylabel('y-value')        # Add y-label to the axes
ax.set_title("Simple plot")     # Add title to the axes

# add grid
ax.grid(True)

# add legend
ax.legend();
```
