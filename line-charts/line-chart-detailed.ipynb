# Detailed line chart

## Import Modules

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-v0_8')
```

## Testdata

```python
# x-data with 200 points
x = np.linspace(0, 5, 200)  

y1 = 10 * x     # curve 1
y2 = 3*x**2     # curve 2
y3 = x**3       # curve 3
```

##  Detailed chart

```python
# create objects figure und axes, size: 9x4 inch
fig, ax = plt.subplots(figsize = (9,4), facecolor = 'ghostwhite')

# plot all curves with labels for legend and the line parameters: color, style, width
ax.plot(x, y1, label = 'linear', color = 'green', linestyle = 'solid',  linewidth = 0.5)
ax.plot(x, y2, label = 'square', color = 'blue',  linestyle = 'dashed', linewidth = 1)
ax.plot(x, y3, label = 'cubic',  color = 'red',   linestyle = 'dotted', linewidth = 2.5)

# title
ax.set_title("Detailed Plot", loc = 'left')      # position left
# ax.set_title("Detailed Plot", loc = 'center')  # center (default)
# ax.set_title("Detailed Plot", loc = 'right')   # right

# background color
ax.set(facecolor = 'whitesmoke')

# x-axis parameter
ax.set_xlabel('x-value')  # label
ax.set_xlim(0, 6)         # min and max values

# x-scale
ax.set_xscale('linear')   # linear
# ax.set_xscale('log')    # logarithmic

# x-ticks
x_ticks = [1,3,5]
ax.set_xticks(x_ticks)

x_tick_labels = ['one', 'three', 'five']
ax.set_xticklabels(x_tick_labels)

# y-axis parameter
ax.set_ylabel('y-value')  # label
ax.set_ylim(0, 140)       # min and max values

# y-scale
ax.set_yscale('linear')  # linear
# ax.set_yscale('log')   # logarithmic

# y-ticks
y_ticks = [0, 25, 50, 75, 100, 125]
ax.set_yticks(y_ticks)

# display grid
ax.grid(True, color = 'black', linestyle = 'dashed', linewidth = 0.6) # visible
# ax.grid(False) # not visible (default)

# display legend
ax.legend(loc = 'upper left', ncols = 3, labelcolor = 'blue', facecolor = 'white', edgecolor = 'black');
```

## Simple parameterization
With `ax.set` many parameters can be grouped:

```python
# create objects figure und axes, size: 9x4 inch
fig, ax = plt.subplots(figsize = (9,4), facecolor = 'ghostwhite')

# plot all curves with labels for legend and the line parameters: color, style, width
ax.plot(x, y1, label = 'linear', color = 'green', linestyle = 'solid',  linewidth = 0.5)
ax.plot(x, y2, label = 'square', color = 'blue',  linestyle = 'dashed', linewidth = 1)
ax.plot(x, y3, label = 'cubic',  color = 'red',   linestyle = 'dotted', linewidth = 2.5)

# title
ax.set_title("Detailed chart", loc = 'right')

# axis parameters
x_ticks = [1,3,5]
x_tick_labels = ['one', 'three', 'five']
y_ticks = [0, 25, 50, 75, 100, 125]

ax.set(xlim = (0,6),    xlabel = 'x-value', xscale = 'linear', xticks = x_ticks, xticklabels = x_tick_labels,
       ylim = (0, 140), ylabel = 'y-value', yscale = 'linear', yticks = y_ticks, facecolor = 'whitesmoke')

# display grid
ax.grid(True, color = 'black', linestyle = 'dashed', linewidth = 0.6)
# ax.grid(False) # nicht sichtbar

# display legend
ax.legend(loc = 'upper left', ncols = 3, labelcolor = 'blue', facecolor = 'white', edgecolor = 'black');
```
