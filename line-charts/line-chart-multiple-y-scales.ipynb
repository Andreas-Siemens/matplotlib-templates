# Line chart with multiple y-scales

## Import Modules

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Testdata

```python
# x-data with 200 points
x = np.linspace(0, 5, 200)  

y1 = 10 * x     # curve 1
y2 = 3*x**2     # curve 2
y3 = x**3       # curve 3
```

## Chart with 3 y-scales

```python
# create objects figure und axes, size: 9x4 inch
fig, ax1 = plt.subplots(figsize = (9,4))

# plot all curves with labels for legend
ax1.set_title("Simple plot")     # Add title to the axes
ax1.set_xlabel('x-value')        # Add x-label to the axes

# first axes
col1 = 'tab:blue'   # line color
p1, = ax1.plot(x, y1, label = 'linear', color = col1)
ax1.set_ylabel('y-linear')        # Add y-label to the axes
ax1.yaxis.label.set_color(col1)
ax1.tick_params(axis = 'y', colors = col1)

# instantiate a second axes that shares the same x-axis
ax2 = ax1.twinx()

col2 = 'tab:green'  # line color
p2, = ax2.plot(x, y2, label = 'square', color = col2)
ax2.set_ylabel('y-square')        # Add y-label to the axes
ax2.yaxis.label.set_color(col2)
ax2.tick_params(axis = 'y', colors = col2)

# instantiate a third axes that shares the same x-axis
ax3 = ax1.twinx()

col3 = 'tab:red'    # line color
p3, = ax3.plot(x, y3, label = 'cubic', color = col3)
ax3.set_ylabel('y-cubic')        # Add y-label to the axes
ax3.yaxis.label.set_color(col3)
ax3.tick_params(axis = 'y', colors = col3)

# Offset the right spine of ax3
ax3.spines.right.set_position(("axes", 1.1))

# add grid
ax1.grid(True)

# add legend
ax1.legend(loc = 'upper left', ncols = 3, handles=[p1, p2, p3]);
```
